## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
### 2(a). Purchasing Analysis (Total)
#   Find the Number of Unique Items, Average Price, Total Purchase and Total Revenue
unique_items = purchase_data_df['Item ID'].nunique()
average_price = purchase_data_df['Price'].mean()
total_purchase = purchase_data_df['Purchase ID'].count()
total_revenue = purchase_data_df['Price'].sum()
```

```python
# Create a Dictionary from the data calculated fro above cell
purchase_analysis_dict = { 'Number of Unique Items': unique_items
                          ,'Average Price': average_price
                          ,'Number of Purchases': total_purchase
                          ,'Total Revenue': total_revenue
}
#Create the DataFrame from the Dictionary
purchase_analysis_df = pd.DataFrame(purchase_analysis_dict,index=[0])
```

```python
### 2(b). Purchasing Analysis (Total) - Formatted display
# Format being set for the 2 columns of the DataFrame for display with Style
format_dict = { 'Average Price':'${:,.02f}'
               ,'Total Revenue':'${:,.02f}'
}
# Apply the style
(purchase_analysis_df                   # DataFrame to be displayed
 .head()                                # Display the head item of the DataFrame
 .style                                 # Invoke style attribute
 .format(format_dict)                   # Apply the style format for the columns
 .set_caption('Purchasing Analysis (Total)')
 .hide_index()
) 
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
### Purchasing Analysis (Gender)
purchase_gender_df = purchase_data_df.filter(['Gender','SN'], axis=1)
```

```python
# Remove the duplicate data by Gender / Screen Name ('SN') from the DataFrame
purchase_gender_grpby_df = purchase_gender_df.groupby(['Gender'])['SN'].value_counts()

# Calculate the basic data for the counts
count_by_gender = purchase_gender_grpby_df.groupby(['Gender']).count()
total_count_gender = purchase_gender_grpby_df.count()
gender_demographics_df = count_by_gender.to_frame('Total Count')
gender_demographics_df['Percentage of Players'] = (count_by_gender/total_count_gender)
```

```python
# Set the format for the column(s) of the DataFrame for display
format_dict = {'Percentage of Players':'{:,.02%}'}

# Apply the format to the DataFrame
(gender_demographics_df                 # DataFrame to be displayed
 .sort_values( by=['Percentage of Players']
              ,ascending=False
             )                          # Sort the DataFrame in Ascending order of the specified column
 .head()                                # Display the head item of the DataFrame
 .style                                 # Invoke style attribute
 .format(format_dict)                   # Apply the style format for the columns
 .highlight_max(color='lightgreen')     # Highlight the highest value data
 .highlight_min(color='#cd4f39')        # Highlight the lowest value data
 .set_caption('Gender Demographics')
)
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Create a new copy of the origianl DataFrame and then proced with the next ask.
purchase_age_df = purchase_data_df.copy()
```

```python
# Analyse the data in preparation for Binning
ages = sorted(purchase_age_df.loc[:, 'Age'].unique())
len(ages)
max(ages)
min(ages)
print(f'Age stats:\n \
        Number of unique Ages: {len(ages):2}\n \
                      Max Age: {max(ages):2}\n \
                      Min Age: {min(ages):2}')
```

```python
#create 9 bins and 8 groups for age data, min_age = 7 and max_age = 45
# refer to the analysis in the above cell
age_bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 50]
age_group = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']
purchase_age_df['Age Ranges'] = pd.cut(purchase_age_df['Age'], bins=age_bins, labels=age_group, include_lowest=True)
purchase_age_df.set_index('Age Ranges', inplace=True)
purchase_age_df.tail(10)
```

```python
# Calculate the values from the DataFrame grouped by 'Age Ranges'
purchase_count_age = purchase_age_df.groupby(['Age Ranges'])['Purchase ID'].count()  # purchase_count
average_purchase_price_age = purchase_age_df.groupby(['Age Ranges'])['Price'].mean() # average_purchase_price
total_purchase_value_age = purchase_age_df.groupby(['Age Ranges'])['Price'].sum()    # total_purchase_value
avg_total_purchase_pp_age = total_purchase_value_age/ \
                            purchase_age_df.groupby(['Age Ranges'])['SN'].nunique()  # avg_total_purchase_pp
```

```python
# Create a new copy of the origianl DataFrame and then proced with the Top Spender analysis
purchase_top_df = purchase_data_df.copy()
purchase_top_df.tail()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Create a filtered copy of the original DataFrame for the analysis of the 'Most Popular Item'
purchase_pop_df = purchase_data_df.filter(['Item ID', 'Item Name', 'Price'], axis=1)
```

```python
# Calculate the values from the DataFrame grouped by 'Item ID', 'Item Name'
purchase_count_pop = purchase_pop_df.groupby(['Item ID'])['Item Name'].value_counts()      # purchase_count
item_price_pop = purchase_pop_df.groupby(['Item ID','Item Name'])['Price'].mean()          # item_price
total_purchase_value_pop = purchase_pop_df.groupby(['Item ID','Item Name'])['Price'].sum() # total_purchase_value
```

```python
# Create a DataFrame from the list
purchasing_pop_df = purchase_count_pop.to_frame('Purchase Count')
purchasing_pop_df['Item Price'] = item_price_pop
purchasing_pop_df['Total Purchase Value'] = total_purchase_value_pop
```

```python
# Sort the DataFrame and apply the data display style to format the data

# Format being set for the 3 columns of the DataFrame for display
format_dict = { 'Purchase Count': '{:,}'
               ,'Item Price':'${:,.02f}'
               ,'Total Purchase Value':'${:,.02f}'
}

(purchasing_pop_df                      # DataFrame to be displayed
 .sort_values( by=['Purchase Count']
              ,ascending=False
             )                          # Sort the DataFrame in Ascending order of the specified column
 .head()                                # Display the head item of the DataFrame
 .style                                 # Invoke style attribute
 .format(format_dict)                   # Apply the style format for the columns
 .highlight_max(color='lightgreen')     # Highlight the highest value data
 .highlight_min(color='#cd4f39')        # Highlight the lowest value data
 .set_caption('Most Popular Item')
) #.hide_index()
```
